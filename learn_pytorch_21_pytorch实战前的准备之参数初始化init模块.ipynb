### 1 参数初始化
init模块位置：torch.nn包中，from torch.nn import init  
官网：https://pytorch.org/docs/stable/nn.init.html   

### 2 初始化方法介绍 

```python

from torch.nn import init
import torch

w = torch.empty(3, 3)

uniform = init.uniform_(w)
print(uniform)

# 1 
normal = init.normal_(w)
print(normal)

# 2 
constant = init.constant_(w, 2.0)
print(normal)

# 3  
eye = init.eye_(w)
print(eye)

# 4   
w_ = torch.empty(3, 16, 5, 5)
dirac = init.dirac_(w_)
print(dirac)

# 5 
xavier_uniform = init.xavier_uniform_(w)
print(xavier_uniform)

# 6   
xavier_normal = init.xavier_normal_(w)
print(xavier_normal)

# 7    
kaiming_uniform = init.kaiming_uniform_(w)
print(kaiming_uniform)

# 8   
kaiming_normal = init.kaiming_normal_(w)
print(kaiming_uniform)

# 9   
orthogonal = init.orthogonal_(w)
print(orthogonal)

# 10    
sparse = init.sparse_(w, sparsity=0.2)
print(sparse)
```
